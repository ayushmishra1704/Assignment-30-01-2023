```python
"""Q.2. Scatter plot is a Matplotlib visualisation tool which does not plot any graph on axes as as continous line  instead represent them as dots (at some particular distance from x and y axes).
"""
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("xaxis"
plt.ylabel("yaxis")
plt.title("Scatter_plot")
```

```python
"""Q.3. Subplot() is a way to create multiple plots within the same figure.
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5]) 
y = (np.array([0, 100, 200, 300, 400, 500]), np.array([50, 20, 40, 20, 60, 70]), np.array([10, 20, 30, 40, 50, 60]), np.array([200, 350, 250, 550, 450, 150]))

f, (ax) = plt.subplots(1,1)
for i in y:
    ax.plot(x, i)
    ax.set_title('This is my fig')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5]) 
y = [np.array([0, 100, 200, 300, 400, 500]), np.array([50, 20, 40, 20, 60, 70]), np.array([10, 20, 30, 40, 50, 60]), np.array([200, 350, 250, 550, 450, 150])]
f, axes = plt.subplots(1,4, figsize=(10, 8))
for i,j in enumerate(axes):
    j.plot(x, y[i])
    j.set_title(f'line {i+1}')
```

```python
"""Q.4. Bar plot is a visualisation tool of matplotlib library which is used to create rectangular(bar) plot on x,y axes.
Bar plot is used for comparision in Data Analysis and  statistics.
Bar plot is used to represent relationship between variables of categorical data.
Bar plot is used to represent sales amount, profit and loss for categorical data.
"""
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
Barplot = plt.bar(company, profit)
Barplot
```

```python
Horizontal_Barplot = plt.barh(company, profit)
```

```python
"""Q.5. Box plot is a visualisation tool used to study the insights of the dataset about central tendency(median). 
Box plot is used in data preprocessing to detect outliers.
Box plot is used to compare distributions between different groups or datasets.
"""
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
```
