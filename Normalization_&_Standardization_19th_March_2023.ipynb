```python
"""Q.1. It is a normalization technique used to scale any numerical feature between 0 and 1.
If the numerical feature has data in the high range but it is not interpretable, or cannnot be 
used in ML models and algorithms, it needs to be scaled down to know the relevance.
It ensures all feature values  are on similar scale preventing feature with larger scale from 
dominating those with smaller scale.
It is used in Deep Learning.
Here is the working model-:
1. Find the datapoints
2. Find the min value and max value from the data points
3. Apply the formula x_scaled = (x - x_max)/(x_max-x_min).
4. New scaled value are in the range 0 to 1.
"""
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
encoder = MinMaxScaler()
```

```python
encoded_df = encoder.fit_transform(df[['total_bill']])
```

```python
df['MinMaxScaler'] = encoded_df
```

```python
df.head()
```

```python
"""Q.2. It is also a normalization technique used to scale the values of  individual samples(rows)
in a dataset to have a vector normal of 1.
It ensures each sample has same magnitude.
It interpret the datapoints in terms of vector.
It is useful when direction of datapoints is more important than magnitude.
Min Max scaling scales the numerical feature in the range of 0 to 1 whereas 
Unit vector scaling scales the values of samples to have a vector normal of 1.
Min max scaling is done for individual datapoinnts of a feature, where as 
Unit vector scaling is done for individual  samples in a dataset.
"""
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
"""Q.3. Principal Component  Analysis is an statistical unsupervised learning method which is als oused in data 
preprocessing. It is basically used for dimensionality reduction, i.e. reducing 2-dimension to
1-dimension.
It is used in face recognition, pattern recognition and once we have found pattern in the data we
can compress data.
Steps for PCA
1. Get some data, let say it contains variables x and y.
2. Calculate the mean of x and y, x_mean and y_mean.
3. Subtract x_mean and y_mean from every data point of x and y.
4. Calculate the covariance matrix of x and y
5. Calculate the eigen vector and eigen values
6. Choose the eigen vector with highest eigen values and form a feature vector.
7. Derive the new dataset, by multiplying transpose of feature vector in matrix form with transpose of x and y column.
"""
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[1,-2], [2,-3], [-1,3], [2,-3], [-1,-8]])
pca = PCA(n_components = 2)
pca.fit(X)
print("explained_variance_ratio",pca.explained_variance_ratio_)
print("singular_values",pca.singular_values_)
```

```python
"""Q.4. Feature extraction is basicaly selecting and extracting most important feature from raw data, 
and PCA is also dimensionality reducing technique by calculations of covariance and eigen vectors and 
finding feature vector and transformed data
"""
import pandas as pd
df = pd.DataFrame({
    'height' : [165,163,159,161,160],
    'weight' : [71, 69, 68, 66, 67],
    'age' : [27, 29, 28, 26, 25],
    'gender' : ['M', 'F', 'M', 'F', 'F'],
    'Blood_pressure' : [128, 125, 95, 109, 121] 
})
from sklearn.decomposition import PCA
X = np.array([df['height'],df['weight'],df['age']])
pca = PCA(n_components = 3)
pca.fit(X)
print("explained_variance_ratio",pca.explained_variance_ratio_)
print("singular_values",pca.singular_values_)
```
