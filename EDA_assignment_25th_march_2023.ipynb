```python
"""Q.13.""" 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv")
```

```python
df.head(2)
```

```python
df["Price"].unique()
```

```python
df["Price"] = df["Price"].str.replace("$", "")
df["Price"] = df["Price"].str.replace("Everyone", "0")
```

```python
df["Price"] = df["Price"].dropna(how = "any")
```

```python
df["Price"] = df["Price"].astype(float)
```

```python
df["Price"].unique()
```

```python
df["Average_Price"] = np.average(df["Price"])
```

```python
fig,axis = plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.barplot(x= df["Type"], y = df["Price"], data = df)
plt.subplot(122)
sns.barplot(x= df["Type"], y = df["Average_Price"], data = df)
```

```python
"""Apps which are paid are distributed around average price of 12.5 to 15.0
Average price is 1."""
```
