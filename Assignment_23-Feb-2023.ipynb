```python
"""Q.5. There are different types of common functions which we manipulate data in Pandas Dataframe:
1. Join
2. Merge
3. set_index
4. reset_index"""
df = pd.DataFrame({'Key1': [2,3,4,5,6],
         'Key2': [3,4,5,6,7],
         'Key3': [4,5,6,7,8]},
          index = ['a','b','c','d','e'])
df
```

```python
df1 = pd.DataFrame({'Key2': [2,3,4,57,67],
         'Key1': [3,4,521,62,71],
         'Key6': [4,52,64,76,81]},
         index = ['a','b','h','i','j'])
df1
```

```python
pd.merge(df,df1,how="left")
```

```python
pd.merge(df,df1,how="right")
```
