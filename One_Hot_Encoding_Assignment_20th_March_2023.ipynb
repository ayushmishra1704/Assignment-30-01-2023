```python
"""Q.1. Data encoding is a technique used to convert categorical variable into
numerical variable that can be used in data preprocessing and used to train ML models.
Categorical data represents qualitative variables such as color, types which doesnt have numerical 
representation
They are used in ML models as input for algorithms such as-
1. Regression
2. Classification
3. Preprocessing like MICE,SICE, Data imputation
"""
```

```python
"""Q.2. Nominal Encoding is a data encoding technique which is used to convert nominal variables
(which has no ordering and ranking of variables) into numerical variables that can be 
further used in ML algorithms.
One common method of Nominal Encoding is One Hot Encoding.
"""
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['sex']])
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded.toarray(),columns = encoder.get_feature_names_out())
```

```python
pd.concat([df,encoded_df], axis = 1)
```

```python
"""Q.3. One hot encoding is one of the method of nominal encoding which creates a binary 
vector for each variable. 
Nominal Encoding is prefered in following situations-:
1. One hot encoding increases the number of variables as it represents
each categorical variables values as a new variable. In such situation we can prefer Nominal
encoding as it represents each category as an integer.
2. One Hot encoding consumes lot of memory space as it represents each feature value as a 
new feature.
3. Nominal encoding can be used when we have to deal with imbalanced dataset.
4. In some cases maintaining the interpretability of model is important. One hot encoding makes 
the interpretation of model less important due to large number of binary features generated.
Nominal encoding preserve the original categorical values
One practical example where label encoding is done is given by-"""
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoded = encoder.fit_transform(df['sex'])
```

```python
df['encoded_sex'] = encoded
```

```python
df.head()
```

```python
"""Q.7. I would use OneHot encoding as number of attribute values of categorical variables are less, so less number of features will be generated. 
"""
# prepare a dataset
import pandas as pd
df = pd.DataFrame({
    'customer_gender' : ['Male','Female','Male', 'Female', 'Female', 'Male'],
    'age' : [25,28,27,32,29,26],
    'contract_type' : ['Permanent', 'Temporary', 'Temporary', 'Permanent', 'Temporary', 'Permanent' ],
    'monthly_charges' : [2500,1500,2600,2400,2100,2500],
    'tenure' : [1,2,3,4,5,6]
})
```

```python
df.head()
```

```python
# use onehotencoding to do encoding as  there are less number of features
df.head()
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
# transform the variable into numerical variable 
encoded = encoder.fit_transform(df[['customer_gender']])
```

```python
# prepare a new dataframe to add the encoded variables.
encoded_df = pd.DataFrame(encoded.toarray(), columns = encoder.get_feature_names_out())
```

```python
pd.concat([df,encoded_df])
```
