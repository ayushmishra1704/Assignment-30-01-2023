```python
"""Q.2. """
import pandas as pd
def reindex():
    df = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8],'C':[9,10,11,12]})
    df1 = df.reindex([x for x in range(1,len(df),2)])
    return df1
reindex()
```

```python
"""Q.3. """
import pandas as pd
def Sum_of_Values():
    df = pd.DataFrame({'Values':[10,20,30,40,50]})
    sum = 0
    for i in range(0,3):
        sum = sum + df['Values'][i]
    return sum
Sum_of_Values()
```

```python
"""Q.7."""
import pandas as pd
def filter_email():
    df = pd.DataFrame({'Email': ["xyz@yahoo.com", "abc@yahoo.com", "pqr@google.com"]})
    df['Username'] = df['Email'].apply(lambda x : x.split("@"))
    df['Username'] = df['Username'].apply(lambda x : x[0])
    return df
filter_email()
```

```python
"Q.9."
import pandas as pd
def Central_Tendency(df):
    df = pd.DataFrame({"Values":[31,82,36,29,93,12,23]})
    Mean = df.mean()
    Median = df.median()
    Mode = df.mode()
    return ("Mean", Mean), ("Median", Median), ("Mode", Mode)
Central_Tendency(df)
```

```python
"Q.10."
import pandas as pd
def calculate_moving_average(a):
    df = pd.DataFrame({'Sales':[450000,350000,475000,540000,240000,980000,780000], 'Date':['2023-11-25', '2023-11-26', '2023-11-27', '2023-11-28', '2023-11-29','2023-11-30','2023-11-01']})
    df['MovingAverage'] = df['Sales'].rolling(window=7).mean()
    return df
calculate_moving_average(df)
```

```python
"Q.11." 
import pandas as pd
def Weekday():
    df = pd.DataFrame({'Date':['2023-01-01', '2023-01-02', '2023-01-03', '2023-11-04', '2023-01-05']})
    df['Weekday'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Weekday'].dt.day_name()
    return df
Weekday()
```

```python
"Q.12." 
import pandas as pd
def datetime():
    date = pd.date_range(start = '2023-01-01 08:00:00', end = '2023-02-01 05:45:00')
    df = pd.DataFrame({'Date':date})
    df['Date'] = pd.to_datetime(df['Date'])
    return(df[1:])
datetime()
```
