```python
"""Q.1. Label Encoding encodes the categorical variable wrt alphabetical order 
of values, wheres in ordinal variable order(category) needs to be defined for encoding
of variables.
"""
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoded_df = encoder.fit_transform(df['sex'])
```

```python
df['sex_encoded'] = encoded_df
```

```python
df.head()
```

```python
df.groupby('class')['fare'].mean()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories = [['First','Second','Third']])
```

```python
encoded_df = encoder.fit_transform(df[['class']])
```

```python
df['encoded_class'] = encoded_df
```

```python
df.head()
```

```python
"""Q.2. Target Guided Ordinal encoding is used when we have too many unique categories 
in the categorical variable . It works based on the relationship with target varriable.
Mean, Median or mode of the target variable replaces categories of the 
categorical variable.
"""
mean_age = df.groupby('sex')['pclass'].mean().to_dict()
```

```python
df['sex_encoded'] = df['sex'].map(mean_age)
```

```python
df.head()
```

```python
"""Q.7. 
"""
import pandas as pd
df = pd.DataFrame({
'Weather_Condition' : ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy', 'Rainy'],
'Wind_Direction' : ['North','South','East','West', 'North','South','East','West', 'North'],   
'Temperature' : [25, 22, 18, 28,20, 26, 23, 19, 30],
'Humidity' : [60, 75, 80, 50, 90, 55, 70, 85, 45]
})
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['Weather_Condition','Wind_Direction']])
```

```python
encoded_df = pd.DataFrame(encoded.toarray(), columns = encoder.get_feature_names_out())
```

```python
df_encoded = pd.concat([df,encoded_df])
```

```python
df_encoded.cov()
```
